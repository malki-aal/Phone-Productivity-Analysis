```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use("seaborn-v0_8")

df = pd.read_csv("../data/clean_data.csv")

df.head()
```

```python
print("Dataset shape:", df.shape)

df.describe()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["phone_usage_hours"],
    df["productivity_score"],
    alpha=0.3,
    s=20
)

x = df["phone_usage_hours"]
y = df["productivity_score"]

m, b = np.polyfit(x, y, 1)

plt.plot(x, m*x + b)

plt.xlabel("Phone Usage (hours)")
plt.ylabel("Productivity Score")

plt.title("Phone Usage vs Productivity")

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["phone_usage_hours"],
    df["study_hours_per_day"],
    alpha=0.05,
    s=15
)

x = df["phone_usage_hours"]
y = df["study_hours_per_day"]

m, b = np.polyfit(x, y, 1)

plt.plot(x, m*x + b)

plt.xlabel("Phone Usage (hours)")
plt.ylabel("Study Hours")

plt.title("Phone Usage vs Study Time")

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hexbin(
    df["phone_usage_hours"],
    df["study_hours_per_day"],
    gridsize=40,
    cmap="Blues"
)

plt.colorbar(label="Number of Students")

plt.xlabel("Phone Usage (hours)")
plt.ylabel("Study Hours")

plt.title("Density of Phone Usage vs Study Time")

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["phone_usage_hours"],
    df["sleep_hours"],
    alpha=0.05,
    s=15
)

x = df["phone_usage_hours"]
y = df["sleep_hours"]

m, b = np.polyfit(x, y, 1)

plt.plot(x, m*x + b)

plt.xlabel("Phone Usage (hours)")
plt.ylabel("Sleep Hours")

plt.title("Phone Usage vs Sleep")

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["sleep_hours"],
    df["productivity_score"],
    alpha=0.3,
    s=20
)

x = df["sleep_hours"]
y = df["productivity_score"]

m, b = np.polyfit(x, y, 1)

plt.plot(x, m*x + b)

plt.xlabel("Sleep Hours")
plt.ylabel("Productivity Score")

plt.title("Sleep vs Productivity")

plt.grid(True)
plt.show()
```

```python
print("Phone vs Productivity:", df["phone_usage_hours"].corr(df["productivity_score"]))

print("Phone vs Study:", df["phone_usage_hours"].corr(df["study_hours_per_day"]))

print("Phone vs Sleep:", df["phone_usage_hours"].corr(df["sleep_hours"]))

print("Sleep vs Productivity:", df["sleep_hours"].corr(df["productivity_score"]))
```

```python
high_phone = df[df["phone_usage_hours"] > 5]
low_phone = df[df["phone_usage_hours"] <= 5]

print("High phone usage avg productivity:",
      high_phone["productivity_score"].mean())

print("Low phone usage avg productivity:",
      low_phone["productivity_score"].mean())
```
